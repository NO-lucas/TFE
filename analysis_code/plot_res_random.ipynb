```python
import matplotlib.pyplot as plt

rank_4 = [56.35, 68.59, 74.26, 78.34, 79.37]
rank_8 = [54.76, 74.83, 75.74, 78.91, 81.52]
shots = [2, 5, 10, 50, 100]


plt.figure(figsize=(8, 5))
plt.plot(shots, rank_4, marker='o', linestyle='-', color='b', label='Rank 4')
plt.plot(shots, rank_8, marker='s', linestyle='--', color='r', label='Rank 8')

plt.xlabel("Shots", fontsize=12)
plt.ylabel("Rank", fontsize=12)
plt.title("Rank vs Shots", fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(shots)
plt.yticks(range(50, 90, 5))
plt.show()
```
